```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weatherData.csv')
```

```python
df
```

```python
plt.hist(df["Temp"],color='green',rwidth=0.95,edgecolor='black',bins=30)
plt.xlabel("Temperature")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Temp', hue='Location', multiple='stack',bins=30)
```

```python
sns.displot(data=df,x='Temp',hue= "Location",col="Season",kind="kde",col_wrap=2,fill=True)
```

```python
sns.displot(data=df,x='Temp',hue= "Location",col="Season",col_wrap=2,kde=True)

#kernel density estimate (KDE) plot
```

```python
sns.displot(data=df,x='Temp',hue= "Location",col="Season",kind="ecdf",col_wrap=2)
```

```python
plt.figure(figsize=(14, 8))

# Generate swarm plot:
sns.swarmplot(data=df, x='Temp', y='Season', hue='Location', palette='magma')
```
